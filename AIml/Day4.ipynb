```python
import numpy as np

choices = np.array([
    [5, 4, 3, 76],   # Row 0
    [84, 5, 4, 3],   # Row 1
    [54, 23, 57, 89] # Row 2
])

# Let's pick:
# Col 0 from Row 1 (84)
# Col 1 from Row 2 (23)
# Col 2 from Row 0 (3)
# Col 3 from Row 1 (3)
indices = [1, 2, 0, 1] 

result = np.choose(indices, choices)
print(result) 
# Output: [84, 23, 3, 3]
```
