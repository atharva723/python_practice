# try catch 

```python
n=int(input("enter your num"))
try:
    print("division by 8 is:{}".format(8/n))
except:
    print("chall")
```

```python
try:
    5+10
except:
    print("give proper i/p")
else:
    print('okay')
```

```python

try:
    n=int(input("enter your num"))
    print("division by 8 is:{}".format(8/n)) 
except:
    print("chall") #executes when any error in try block
```

```python
try:
    n=int(input("enter the num"))
    print("division of {} by 8 is:{}".format(n,int(8/n)))
except ZeroDivisionError as ze:
    print(ze.args)
except TypeError as t:
    print(t.args)
except Execption as e:
    print(e.args)
finally:
    print("this will execute irespective of anything")
    
```
