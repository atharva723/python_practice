```python
import numpy as np
import pandas as pd
import tensorflow
from tensorflow import keras
import matplotlib.pyplot as plt
from tensorflow.keras import layers
```

```python
from keras.datasets import mnist, fashion_mnist, cifar10
```

```python
# m = mnist.load_data()
c = cifar10.load_data()
# mnist.load_data()
```

```python
#(x_train,y_train),(x_test,y_test) = mnist.load_data()
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
```

```python
for i in range(9):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i], cmap=plt.get_cmap('gray'))
    plt.show()
```

```python
print(x_train.shape, " ", x_train.shape[0])
print(x_test.shape," ",x_test.shape[0])
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
train_norm = x_train.astype('float32')
test_norm = x_test.astype('float32')

train_norm = train_norm/255.0
test_norm = test_norm/255.0
```

```python
from keras.models import Sequential
from keras.layers import Dense , Dropout, BatchNormalization
from keras.layers import Conv1D,Conv2D, Conv3D, MaxPooling2D, Flatten
```

```python
model = Sequential()
```

```python
model.add(Conv2D(32,(3,3), activation='relu', input_shape=(32,32,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
```

```python
model.add(Conv2D(64,(3,3), activation='relu', input_shape=(32,32,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))
```

```python
model.add(Conv2D(64,(3,3), activation='relu', input_shape=(32,32,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))

model.add(Flatten())
```

```python
model.add(Dense(100,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
m2 = model.fit(train_norm,y_train, epochs=10, batch_size=25,validation_data=(test_norm,y_test))
```

```python
pred = model.predict(test_norm)
```

```python
from sklearn.metrics import mean_squared_error, confusion_matrix
```

```python
pred[0]
np.argmax(pred[0])
y_test[0]

model.summary()
```

```python
acc=m2.history['accuracy']
val_acc=m2.history['val_accuracy']

print(acc," ",val_acc)
```

```python
loss=m2.history['loss']
val_loss=m2.history['val_loss']

print(loss," ",val_loss)
```

```python
epochs_range = range(10)
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
```

```python
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.savefig('./foo.png')
plt.show()
```
