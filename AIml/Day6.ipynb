## inheritance

```python
class parent:
    sir_name='Smith'
    _age=27
    
```

```python
x= parent()
```

```python
x.sir_name
```

```python
class child(parent):
    Bgroup="AB+"
    _city='Mumbai'
    __gender='male'
```

```python
y=child()
```

```python
y.sir_name
```

```python
x._age
```

```python
y.__gender
```

```python
class child2(child):
    email='atharvajadhav20@'
    phone=992001
```

```python
z=child2()
```

```python
z.email
```

```python
z._age
```

```python
#multiple inheritance
class parent1():
    sir_name='cathrene'
class Multi(parent,parent1):
    pass
```

```python
m=Multi()
```

```python
m.sir_name
```

```python
#encapsulation
class car:
    def __init__(self):
        self._engine_num='En233e'
        self.__engine_no='233e'
```

```python
car.__engine_no
```

```python
class parent:
    def __init__(self,name):
        self.name=name
        self.__machine_code='QJH778'#this will not be inherited 
class child(parent):
    def __init__(self,name):
        super().__init__(name)
        self.__machine_manu='Toyota'
```

```python
p=parent(name='honda')
```

```python
p._parent__machine_code
```

```python
p=child('suzu')
```

```python
p.name
```
