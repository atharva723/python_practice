## Handling Outliers
###--> either selete them if they are anomalies or change the values to normal range or provide additional info to model.
## to tackle outlier we can penalize the model.(we introduce errors to avoid overfitting through penalizing )
Handling Outliers
Outliers are the "weird" data pointslike a 2-room shack in Boston selling for $1M. They affect models differently:

In Linear Regression: Outliers are dangerous. Because the model tries to minimize the squared distance of all points from the line, one extreme outlier can "pull" the entire line toward it, ruining the accuracy for everyone else.

In Decision Trees: Outliers are less impactful. A tree might simply create a tiny, isolated "leaf" for that one weird house. While it doesn't ruin the rest of the tree, it can lead to overfitting (the model memorizing noise instead of patterns).

How to handle them:

Trimming/Capping: Remove the top/bottom 1% of data or "cap" values at a certain threshold.

Transformation: Use np.log(df['Price']) to shrink the distance between extreme values.

## Decision tree regressor
What is a Decision Tree Regressor?
Imagine you are playing a game of 20 Questions to guess the price of a house.

Question 1: "Is the number of rooms greater than 6.5?"

Question 2: "Is the crime rate lower than 0.5%?"

A Decision Tree Regressor does exactly this. It splits your data into smaller and smaller subsets (nodes) based on feature thresholds. 
Instead of a straight line, it creates a step function. In the "leaves" (the bottom of the tree), 
the prediction is simply the average price of all houses that landed in that specific bucket.

```python
from dtreeviz import model
from sklearn.tree import DecisionTreeRegressor

# 1. Train your model (keep max_depth small for a clean visual)
regr = DecisionTreeRegressor(max_depth=5,min_samples_leaf=5,min_samples_split=5)
dt_model=regr.fit(x_train, y_train)

```

# Ensemble models---> bagging (Random Forest),boosting
1. Bagging (Bootstrap Aggregating)
The Concept: Parallel voting.
In Bagging, you create multiple versions of your dataset by "sampling with replacement" (taking random subsets). 
You train a separate tree on each subset simultaneously. When it's time to predict, you take the average of all their answers.

Goal: To reduce Variance (overfitting).

Analogy: Asking 100 people the same question independently and averaging their answers to cancel out the "crazy" outliers.

2. Random Forest
The Concept: Bagging + Feature Randomness.
A Random Forest is just a specialized version of Bagging. 
Not only does it pick random rows of data for each tree, but it also picks a random subset of features (columns) at every split.

Boosting
The Concept: Sequential learning.
Unlike Bagging, where trees grow side-by-side, Boosting trees grow one after another. Each new tree specifically looks at the errors (residuals) of the tree before it.

How it works: Tree 1 makes a prediction. Tree 2 is then trained specifically to predict the mistakes Tree 1 made. Tree 3 tries to fix the combined mistakes of Tree 1 and 2.

Goal: To reduce Bias (underfitting) and improve accuracy.

Popular Algorithms: XGBoost, LightGBM, and Gradient Boosting.
