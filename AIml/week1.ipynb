```python
class SensorAnalyzer:
    def sensor_array(self,sensor_value=list()):
        return np.array(sensor_value,dtype=float)

    def valid_sensor_array(self,sensor_array:np.ndarray):
        if sensor_array.size==0 or np.any(sensor_array<=0) :
            return False
        else:
            return True
    def sensor_statistics(self, sensor_array: np.ndarray):
        m=float(sensor_array.sum())
        l=float(sensor_array.min())
        avg=round(sensor_array.mean())
        return (m,avg,l)
    def filter_extreme(self, sensor_array: np.ndarray):
        ua=sensor_array.astype('float')
        ua[ua>=50.0]=ua[ua>=50.0]*0.90
        return ua
    def label_high_sensor(self, sensor_array: np.ndarray):
        result=np.empty_like(sensor_array,dtype=object)
        for i in range(len(sensor_array)):
            if sensor_array[i] > sensor_array.mean():
                result[i]='High'
            elif sensor_array[i] < sensor_array.mean():
                    result[i]='Low'
            else:
                result[i] ='Medium'
        return result
    def format_sensor_read(self, sensor_array: np.ndarray):
        rs=np.empty_like(sensor_array,dtype=object)
        for i in range(len(sensor_array)):
            rs[i]=f"{sensor_array[i]} units"
        return rs
            
            
                
                
            
A=SensorAnalyzer()

A.format_sensor_read(np.array([1,2,3]))
```

```python
import pandas as pd

class FlightDelayAnalyzer:
    def create_flight_log_df(Self, flight_log: list):
        return pd.DataFrame(flight_log,columns=['FlightID','AirlineCode','Route','Delay'])
    
    def create_airline_master_df(Self, airline_master: list):
        return pd.DataFrame(airline_master,columns= ["AirlineCode", "AirlineName"])
    
    def merge_airline_names(Self, flight_log_df: pd.DataFrame, airline_master_df: pd.DataFrame):
        return pd.merge(flight_log,airline_master_df,on='AirlineCode', how='left')

    def average_delay_by_airline(Self, merged_df: pd.DataFrame):
        return merged_df.groupby('AirlineName')['Delay'].mean().reset_index().rename(columns={'Delay':'Average Delay'})
        
```
