```python
from typing import TypedDict, List, Literal

class myaccount(TypedDict):
    amt_usd = float
    total_usd = float
    total = float
    target_cur = Literal["INR","EUR"]
```

```python
def calc_totamt(state:myaccount) -> myaccount:
    state['total_usd'] = state['amt_usd']*1.08
    return state
```

```python
def convert_toinr(state:myaccount) -> myaccount:
    state['total'] = state['total_usd']*88
    return state
```

```python
def convert_toeur(state:myaccount) -> myaccount:
    state['total'] = state['total_usd']*0.8
    return state
```

```python
def choose_conversion(state:myaccount)->str:
    return state['target_cur']
```

```python
from langgraph.graph import StateGraph, START, END
```

```python
builder = StateGraph(myaccount)
builder.add_node("total_cal_node",calc_totamt)
builder.add_node('calc_total_inr',convert_toinr)
builder.add_node('convert_to_eur',convert_toeur)
```

```python
builder.add_edge(START, 'total_cal_node')
builder.add_conditional_edges("total_cal_node",
                              choose_conversion,
                              {
                                  "INR":"calc_total_inr",
                                  "EUR":"convert_to_eur"
                              }
                              )
builder.add_edge(['calc_total_inr', 'convert_to_eur'],END)
```

```python
graph = builder.compile()
```

```python
display(Image(graph.get_graph().draw_mermaid_png()))
```
