```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price($)')
plt.scatter(df.price,df.area,marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
d = pd.read_csv('area.csv')
d.head(3)
reg.predict(d)
```

```python
p = reg.predict(d)
```

```python
d['price'] =p
```

```python
d
```

```python
d.to_csv('predict.csv')
```

```python
d = pd.read_csv('predict.csv',index_col=0)
d
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price($)')
plt.scatter(df.price,df.area,marker='+')
plt.plot(df.area,reg.predict(df[['area']]))
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'])
```

```python
df.columns
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
df= pd.read_csv('year.csv')
df
```

```python
p = reg.predict(df)
```

```python
df['per capita income (US$)'] = p
```

```python
df
```

```python
df.to_csv('new_canada.csv')
df
```

```python
df.columns
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='green',marker='*')
plt.plot(df['year'], reg.predict(df[['year']]), color='yellow')
```
