```python
import pandas as pd 

data= pd.read_csv(r"C:\Users\Atharva\Desktop\ML\logistic\logistic multi variable\boston_house_prices.csv",header=1)
```

```python
x= data.drop(['MEDV'],axis=1)
```

```python
y= data['MEDV']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=56,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
r_tree= RandomForestRegressor()
```

```python
r_tree.fit(x_train,y_train)
```

```python
r_tree.predict(x_test)
```

```python
import numpy as np

param_distribution = {
    'n_estimators': [int(x) for x in np.linspace(100, 1000, num=10)],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [int(x) for x in np.linspace(10, 110, num=11)],
    'min_samples_split': [2, 5, 10],  # Corrected 'min_sample_split' to 'min_samples_split'
    'min_samples_leaf': [1, 2, 4],    # Corrected 'min_samples_leaf'
    'bootstrap': [True, False]
}
```

# setup randomized search

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,root_mean_squared_error
```

```python
random_search= RandomizedSearchCV(
    estimator=r_tree,
    param_distributions = param_distribution,
    n_iter=100,
    cv=3,
    verbose=2,
    random_state=55,
    n_jobs=-1
)
```

```python
random_search.fit(x_train,y_train)
```

```python
best_param= random_search.best_params_
print("best hyperparam:",best_param)
```

```python
best_model= random_search.best_estimator_
y_pred= best_model.predict(x_test)
```
