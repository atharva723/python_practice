# 1.Tokenization


```python
text = "The sun was setting over the horizon, casting a warm golden glow across the landscape. As the evening shadows lengthened, the cool breeze rustled through the trees, bringing a sense of tranquility. In the distance, the gentle hum of a nearby stream added to the serene ambiance, making it a perfect moment for reflection."
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
word_tokens=word_tokenize(text)
word_tokens
```

```python
sent_tokens=sent_tokenize(text)
sent_tokens
```

```python
text2="Heloo%^ my kejf&&%& is sap $65 in the$^#^ La >>>>>>>>><M<<<<<<<<b"
clean_word_token=word_tokenize(text2)
clean_word_token
```
